# Preprocesamiento de Datos

```python
# Importar librerías
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder

```

```python
# Importar la base datos:
df = pd.read_csv('census.csv')
```

```python
# Realizar dataset:
df.head()
```

```python
# Información de las columnas:
df.info()
df.describe()
```

```python
# Verificar si hay datos faltantes:
print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
# Eliminar las filas que tengan 3 o más datos faltantes:
filasantes = len(df)
df = df[df.isnull().sum(axis=1) < 3]
filasdespues = len(df)
filaseliminadas = filasantes - filasdespues

print(f"Filas antes: {filasantes}")
print(f"Filas después: {filasdespues}")
print(f"Filas eliminadas: {filaseliminadas}")
```

```python
# Contar nuevamente los datos faltantes:
print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
# Imputar los datos faltantes de edad y horas trabajadas por semana con la mediana de cada una de esas columnas:
df['age'] = df['age'].fillna(df['age'].median())
df['hours_per_week'] = df['hours_per_week'].fillna(df['hours_per_week'].median())
```

```python
# Contar de nuevo los datos faltantes:
print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
# Aplicar one-hot encoding a la columna "workclass":
df = pd.get_dummies(df, columns=['workclass'], prefix='workclass')
df.head()
```

```python
# Encontrar los valores que toma la columna "education":
print(df['education'].unique())
print(df['education'].nunique())
```

```python
# Aplicar ordinal encoding a la columna "education":
oe= OrdinalEncoder()
df['education'] = oe.fit_transform(df[['education']])
df.head()
```

```python
# Comprobar que la columna "education" tenga los valores apropiados:
print(df['education'].unique())
print(df['education'].nunique())
print(f"\nRango: {df['education'].min()} a {df['education'].max()}")
```

```python
# Aplicar one-hot encoding a la columna "race":
df = pd.get_dummies(df, columns=['race'], prefix='race')
df.head()
```

```python
# Apliqcar binary encoding a la columna "sex":
le= LabelEncoder()
df["sex"]= le.fit_transform(df["sex"])
df.head()
```

```python
# Aplicar binary encoding a la columna "label":
le= LabelEncoder()
df["label"]= le.fit_transform(df["label"])
df.head()
```

```python
# Veamos la infomación final del dataset:
df.info()
df.describe()

```

```python
# Verificar el tipo de dato de cada columna:
print(df.dtypes)
```

```python
# Cargar el dataset a un nuevo archivo CSV:
df.to_csv('census_preprocessed.csv', index=False)   

```
