```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar la base de datos
df = pd.read_json("pokemonDB_dataset.json")
df = df.T
```

```python
# Mostrar las primeras filas del DataFrame
df.head()
```

```python
# Mostrar las columnas del DataFrame
df.columns
```

```python
# Mostrar información del DataFrame
df.info()
```

```python
# Mostrar estadísticas descriptivas del DataFrame
df.describe(include='all')
```

```python
# Contar los tipos de datos en el DataFrame
df.dtypes.value_counts()
```

```python
# Columnas numéricas
numericas = df.select_dtypes(include=['int64', 'float64']).columns
print("Columnas numéricas:", list(numericas))

# Columnas categóricas
categoricas = df.select_dtypes(include=['object']).columns
print("Columnas categóricas:", list(categoricas))
```

```python
# Limpiar las columnas "Height" y "Weight"
df["Height"] = df["Height"].str.replace(" m", "", regex=False)
df["Weight"] = df["Weight"].str.replace(" kg", "", regex=False)
df["Weight"] = df["Weight"].str.replace(" lbs", "", regex=False)
```

```python
# Definir listas de columnas numéricas
cols_num = ["HP Base", 
        "HP Min",
        "HP Max",
        "Attack Base",
        "Attack Min",
        "Attack Max",
        "Defense Base",
        "Defense Min",
        "Defense Max",
        "Special Attack Base",
        "Special Attack Min",
        "Special Attack Max",
        "Special Defense Base",
        "Special Defense Min",
        "Special Defense Max",
        "Speed Base",
        "Speed Min",
        "Speed Max",
        "Base Exp"]
```

```python
# Convertir columnas a tipo numérico
for i in cols_num:
    df[i]=pd.to_numeric(df[i], errors="coerce")
```

```python
# Convertir "Height" y "Weight" a tipo numérico
df["Height"] = df["Height"].str.extract(r"([0-9]*\.?[0-9]+)").astype(float)
df["Weight"] = df["Weight"].str.extract(r"([0-9]*\.?[0-9]+)").astype(float)
```

```python
# Verificar los tipos de datos después de la conversión
df.dtypes
```

```python
# Mostrar las primeras filas del DataFrame
df.head()
```

```python
# Dividir la columna "Type" en "Type1" y "Type2"
df[["Type1", "Type2"]] = df["Type"].str.split(",", n=1, expand=True)
```

```python
# Mostrar las primeras filas de las nuevas columnas
df[["Type", "Type1", "Type2"]].head()
```

```python
# Dividir las columnas por género
df[['Male (%)', 'Female (%)']] = df['Gender'].str.split(',', expand = True)
```

```python
# Mostrar las primeras filas de las nuevas columnas
df[["Gender", "Male (%)", "Female (%)"]].head()
```

```python
# Verificar valores nulos 
df.isnull().sum()
```

```python
# Rellenar valores nulos 
df["Weight"]=df["Weight"].fillna(df["Weight"].median())
df["Type2"]=df["Type2"].fillna("None")
df["Base Exp"]=df["Base Exp"].fillna(df["Base Exp"].median())
df["Female (%)"]=df["Female (%)"].fillna("0")
```

```python
# Verificar valores nulos después de rellenar
def valores_duros(variable) :
    umbral = 1.5
    q1 = variable.quantile(0.25)
    q3 = variable.quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - umbral * iqr
    limite_superior = q3 + umbral * iqr
    valores = variable[(variable < limite_inferior) | (variable > limite_superior)]
    return valores

valores_duros(df['Weight'])
valores_duros(df['Height'])
valores_duros(df['HP Base'])
valores_duros(df['Attack Base'])
valores_duros(df['Defense Base'])
valores_duros(df['Speed Base'])
valores_duros(df['Base Exp'])
```

```python
# Crear un nuevo DataFrame limpio con las columnas necesarias
df_clean = df[[
    "Type1", "Type2",
    "HP Base", "Attack Base", "Defense Base", "Speed Base",
    "Height", "Weight", "Base Exp", "Species"
]].copy()
```

```python
# 1. Promedio, mínimo y máximo de los atributos base (HP, Attack, Defense, Speed)
atributos_base=["HP Base", "Attack Base", "Defense Base", "Speed Base"]
stats=df_clean[atributos_base].agg(["mean", "min", "max"]).T
stats
```

```python
# 2. Histograma de "Base Exp"
sns.histplot(df_clean["Base Exp"], bins=25, kde=False, color="purple")
plt.title("Distribución de la experiencia base (Base Exp)")
plt.ylabel("Frecuencia")
plt.xlabel("Base Exp")
plt.show()
```

```python
# Tipos de Pokémon principales
df_clean["Type1"].unique()
```

```python
# 3. Boxplot de "Attack Base" por "Type1"
plt.figure(figsize=(14,6))
sns.boxplot(x="Type1", y="Attack Base", data=df_clean, color="purple")
plt.title("Comparación de Attack Base entre tipos principales", fontsize=12)
plt.xlabel("Tipo principal (Type1)")
plt.ylabel("Attack Base")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 6. Correlación entre atributos de base (HP, Attack, Defense, Speed)
correlacion = df_clean[atributos_base].corr()
correlacion
```

```python
# 8. Scatter plot entre "Weight" y "Attack Base"
plt.figure(figsize=(9,7))
sns.scatterplot(x="Weight", y="Attack Base", data=df_clean, alpha=0.7, color="purple")
plt.title("Relación entre peso (Weight_kg) y ataque base (Attack Base)", fontsize=14)
plt.xlabel("Weight (kg)")
plt.ylabel("Attack Base")
plt.show()
```

```python
# 9. Promedio de altura y peso por tipo principal
promedio_peso = df_clean.groupby('Type1')['Weight'].mean()
promedio_altura = df_clean.groupby('Type1')['Height'].mean()
tabla = pd.DataFrame({'Promedio por peso: En Kg' : promedio_peso, 'Promedio por altura: En M' : promedio_altura})
tabla.sort_values('Promedio por peso: En Kg', ascending=False)
```
